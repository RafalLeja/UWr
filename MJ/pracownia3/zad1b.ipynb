```python
def representation(L):
    txt = ' '.join(L)
    input_ids = tokenizer(txt, return_tensors='pt')['input_ids'].to(device)
    output = model(input_ids=input_ids)
    return output.last_hidden_state.detach().cpu().numpy()[0,0,:]

print (representation(['kot']))
```

```python
def simpleEmbeddingStr(w):
    return f'{w} ' + ' '.join([f'{x:.4f}' for i,x in enumerate(representation([w]))])

def weightedEmbeddingStr(w, weight):
    return f'{w} ' + ' '.join([f'{x:.4f}' for i,x in enumerate(representation([w])) * weight])

with open('word_embedings_file.txt', 'w') as f:
    for line in clusters_txt.strip().split('\n'):
        words = line.split(' ')[1:]
        for word in words:
            f.write(simpleEmbeddingStr(word) + '\n')
        # f.write('\n')
```
