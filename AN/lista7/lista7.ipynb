```julia
using Plots

plt = plot(
    xlim = (-1, 1),
    ylim = (-0.12, 0.12),
    legend = true,
    title = "punkty równoodległe",
    marker = 2,
)

function PEven(x, n)
    product = 1
    points = range(-1,stop=1, step=2/n)
    for i in points
        product *= (x - i)
    end    
    return product
end
    
for i in range(start = 4, stop = 20)
    plot!(y -> PEven(y,i), label="n = $i")
end
display(plt)
savefig("rownoodlegle.svg")
```

# Porównanie

węzły Czybyszewa zmniejszają efekt Rungego (odchylenie na końcu przediału)

```julia
using Plots

plt = plot(
    xlim = (-1, 1),
#     ylim = (-0.12, 0.12),
    title = "porównanie",
    legend = true,
    marker = 2,
)

function PT(x, n)
    product = 1
    points = [cos((2k-1)*pi/2(n+1)) for k in range(start=1, length=(n+1))]
    for i in points
        product *= (x - i)
    end    
    return product
end

function PEven(x, n)
    product = 1
    points = range(-1,stop=1, step=2/n)
    for i in points
        product *= (x - i)
    end    
    return product
end

for i in range(start=10, step=1, length=1)
    plot!(y -> PEven(y,i), label="n = $i, równoodległe")
    plot!(y -> PT(y,i), label="n = $i, Czebyszewa")
end
display(plt)
savefig("Czebyszewa.svg")
```
