```julia
using Plots

plt = plot(
    1,
    xlim = (-1.5, 1),
    legend = false,
    marker = 2,
)

function PEven(x, n)
    product = 1
    points = range(-1,stop=1, step=2/n)
    for i in points
        product *= (x - i)
    end    
    return product
end
    
x = range(-2, 2, length=100)

for i in range(start = 4, stop = 6)
    plot!(y -> PEven(y,i))
end
display(plt)
```
