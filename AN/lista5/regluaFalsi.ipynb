# regula falsi

## Idea:
 - bierzemy punkty początkowe a i b, gdzie  $\alpha$ należy do (a, b) oraz f(a)f(b) < 0
 - rysujemy prostą z punktu (a, f(a)) do (b, f(b))
 - znajdujemy miejsce zerowe prostej - x0
 - jeżeli funkcja jest tego samego znaku w punktach x0 i a, ustawiamy a jako x0, i odwrotnie 

## Zalety: 
 - zawsze zbierzna (monotonicznie)
 - jeżeli funkcja jest liniowa, otrzymujemy $\alpha$ w jednym kroku
 - nie potrzeba pochodnej

## Rożnice:
 - przedział zawsze maleje
 - przedział nie jest wybierany z dwóch ostatnich x, tylko z ostatnich x o przeciwnym znaku f(x) 

```julia
using Plots

f(x) = sign(x)

a = -3
b = 2.5

plt = plot(
    1,
    xlim = (-3, 3),
    legend = false,
    marker = 2,
)

plot!(plt, f)
plot!(plt, [a, b], [f(a), f(b)]) 
plot!(plt, [a, a], [0, f(a)], linestyle = :dash) 
plot!(plt, [b, b], [0, f(b)], linestyle = :dash) 
scatter!(plt, [a, a, b, b],[0, f(a), 0, f(b)])
itr = 1
while (itr < 5)
    itr = itr + 1

    x = a - f(a) * (a-b) / (f(a)-f(b))
    if f(x)*f(a) > 0
        a = x
        plot!(plt, [a, a], [0, f(a)], linestyle = :dash) 
    else
        b = x
        plot!(plt, [b, b], [0, f(b)], linestyle = :dash) 
    end
    println(x)
    plot!(plt, [a, b], [f(a), f(b)]) 
    scatter!(plt, [a, a, b, b],[0, f(a), 0, f(b)])
    # scatter!(plt, [x],[0])
end
display(plt)
savefig(plt, "exmpl1.png")
```

```julia
using Plots

f(x) = (x)^2 - 4
# f(x) = (x)^3 - 2x - 5

a = 0.5
b = 3

plt = plot(
    1,
    xlim = (a, b),
    legend = false,
    marker = 2,
)

plot!(plt, f)
plot!(plt, [a, b], [f(a), f(b)]) 
plot!(plt, [a, a], [0, f(a)], linestyle = :dash) 
plot!(plt, [b, b], [0, f(b)], linestyle = :dash) 
scatter!(plt, [a, a, b, b],[0, f(a), 0, f(b)])
itr = 1
x0 = 0
while (itr < 50)
    itr = itr + 1

    x = a - f(a) * (a-b) / (f(a)-f(b))
    if f(x)*f(a) > 0
        a = x
        plot!(plt, [a, a], [0, f(a)], linestyle = :dash) 
    else
        b = x
        plot!(plt, [b, b], [0, f(b)], linestyle = :dash) 
    end
    println(abs(x-2)/abs(x0-2))
    x0 = x
    plot!(plt, [a, b], [f(a), f(b)]) 
    scatter!(plt, [a, a, b, b],[0, f(a), 0, f(b)])
end
display(plt)
savefig(plt, "exmpl3.png")
```
